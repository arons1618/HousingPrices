```python
def full_one_hot_preprocess(x):
    copy = x.copy()
    copy.iloc[:,[3, 26, 59]] = copy.iloc[:,[3, 26, 59]].fillna(0)
    copy.iloc[:,[6, 25, 30, 31, 32, 33, 35, 42, 57, 58, 60, 63, 64, 72, 73, 74]] = copy.iloc[:,[6, 25, 30, 31, 32, 33, 35, 42, 57, 58, 60, 63, 64, 72, 73, 74]].fillna("NA")
    copy["MSSubClass"] = copy["MSSubClass"].apply(str)
    
    return pd.get_dummies(copy)
    
```

```python
full_one_hot_test_data = full_one_hot_preprocess(test)
full_one_hot_clean_data = full_one_hot_preprocess(data)
full_one_hot_clean_data.head()
```

```python
full_one_hot_clean_data.to_csv("full_one_hot_train.csv")
full_one_hot_test_data.to_csv("full_one_hot_test.csv")
```
